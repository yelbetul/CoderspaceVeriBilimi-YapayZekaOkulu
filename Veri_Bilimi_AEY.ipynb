### Yararsz Verinin karlmas ve Dzenlenmesi

### Kayp Veri Sorunu 
* Kayp veri kavram veri ierisinde bulunan rnein herhangi bir stununda bulunan deerin olmamas anlamna gelir. 
* Mesela, aada bulunan veride ikinci index'te bulunan rnein boy ve kilo stunlarnda NaN (Not a Number) yazmaktadr. Bu NaN yazs rnee ait boy ve kilo verisinin olmad ya da kayp olduu anlamna gelir.

```python
veri.head(3)
```

* Kayp veri ile mcadele etmek iin,
    1. Kayp veriye ait rnekler veriden karlabilir.
    1. Kayp veriler asl verinin dalmn bozmayacak ekilde asl verinin ortalama ya da medyan deerlerine gre doldurulabilir.
* Bu veri zerinde alma yaparken, 
    1. Boy ve kilo stununda bulunan kayp veriyiyi etkinlik ortalamasna gre dolduracaz.
    1. Ya stununda bulunan kayp veriyi veri setinin ya ortalamasna gre dolduracaz.
    1. Madalya alamayan sporcular veri setinden karacaz.

### Boy ve Kilo Stunu Kayp Veri Doldurma
* Boy ve kilo stununda bulunan kayp veriyiyi etkinlik ortalamasna gre dolduracaz.

```python
# 1. Boy ve kilo stununda bulunan kayp veriyiyi etkinlik ortalamasna gre dolduracaz.
essiz_etkinlik = pd.unique(veri.etkinlik)
print("Esiz etkinlik says: {}".format(len(essiz_etkinlik)))
essiz_etkinlik[:10]
```

```python
# her bir etkinlii iteratif olarak dola
# etkinlik zelinde boy ve kilo ortalamalarn bul
# etkinlik zelinde boy ve kilo da kayp olan deerlere ortalama boy ve kilo deerlerini eitle

veri_gecici = veri.copy() # gerek veriyi kaybetmemek iin veri_gecici deikeni belirle
boy_kilo_liste = ["boy", "kilo"]
for e in essiz_etkinlik: # etkinlik listesi ierisinde dola
    
    # etkinlik filtresi olutur
    etkinlik_filtre = veri_gecici.etkinlik == e
    # veriyi etkinlie gre filtrele
    veri_filtreli = veri_gecici[etkinlik_filtre]
    
    # boy ve kilo iin etkinlik zelinde ortalama bul
    for s in boy_kilo_liste:
        ortalama = np.round(np.mean(veri_filtreli[s]),2)
        if ~np.isnan(ortalama): # eer etkinlik zelinde ortalama varsa
            veri_filtreli[s] = veri_filtreli[s].fillna(ortalama)
        else: # etkinlik zelinde ortalama yoksa tm veri iin ortalama bul
            tum_veri_ortalamasi = np.round(np.mean(veri[s]),2)
            veri_filtreli[s] = veri_filtreli[s].fillna(tum_veri_ortalamasi)
    # etkinlik zelinde kayp deerleri doldurulmu veriyi veri geiciye eitle            
    veri_gecici[etkinlik_filtre] = veri_filtreli

# kayp deerleri giderilmi geici veriyi gerek veri deikenine eitle
veri = veri_gecici.copy() 
veri.info() # boy ve kilo stunlarnda kayp deer saysna bak
```

### Ya Stunu Kayp Veri Doldurma
* Ya stununda bulunan kayp veriyi veri setinin ya ortalamasna gre dolduracaz.

```python
# 2. Ya stununda bulunan kayp veriyi veri setinin ya ortalamasna gre dolduracaz.
# yas deikeni tanml olmayan rnekleri bul, 
# tilda iareti ile tersini al
# ya deikeni tanml olan rnekleri bulmak iin filtre olutur
yas_ortalamasi = np.round(np.mean(veri.yas),2)
print("Ya ortlamas: {}".format(yas_ortalamasi))
veri["yas"] = veri["yas"].fillna(yas_ortalamasi)
veri.info()
```

### Madalya Alamayan Sporcular Veri Setinden kar
* Madalya alamayan sporcular veri setinden karacaz.

```python
# 3. Madalya alamayan sporcular veri setinden karacaz.
# toplamda 231333 tane rnek iin madalya deikeni tanml deil
madalya_degiskeni = veri["madalya"]
pd.isnull(madalya_degiskeni).sum()
```

```python
# madalya deikeni tanml olmayan rnekleri bul, (NaN) 
# tilda iareti ile tersini al
# madalya deikeni tanml olan rnekleri bulmak iin filtre olutur
madalya_degiskeni_filtresi = ~pd.isnull(madalya_degiskeni)
```

```python
veri = veri[madalya_degiskeni_filtresi]
veri.head(5)
```

```python
# ncelikli olarak bar grafiini elde edeceimiz metodumuzu yazalm.
def plotBar(degisken,n = 5):
    """
        Girdi: Deiken/stun ismi
               n = en nemli 10 esiz deer
        kt: Bar grafii
    """
    veri_ = veri[degisken]
    veri_sayma = veri_.value_counts()
    veri_sayma = veri_sayma[:n]
    plt.figure()
    plt.bar(veri_sayma.index, veri_sayma, color = "orange")
    plt.xticks(veri_sayma.index, veri_sayma.index.values)
    plt.xticks(rotation=45)
    plt.ylabel("Frekans")
    plt.title("Veri Skl - {}".format(degisken))
    plt.show()
    print("{}: \n {}".format(degisken, veri_sayma))
    
```

## Anomali Tespiti
* Aykr deer, dier gzlemlerden nemli lde farkl bir veri noktasdr. Bir aykr deer, lmdeki deikenlie bal olabilir veya deneysel hatay gsterebilir; ikincisi bazen veri kmesinden karlr. Bir aykr deer istatistiksel analizlerde ciddi sorunlara neden olabilir.
* 

```python
def anomaliTespiti(df,ozellik):
    outlier_indices = []
    
    for c in ozellik:
        # 1. eyrek
        Q1 = np.percentile(df[c],25)
        # 3. eyrek
        Q3 = np.percentile(df[c],75)
        # IQR: eyrekler akl
        IQR = Q3 - Q1
        # aykr tespiti iin arpan
        outlier_step = IQR * 1.5
        # aykry ve aykr indeksini tespit et
        outlier_list_col = df[(df[c] < Q1 - outlier_step) | (df[c] > Q3 + outlier_step)].index
        # indeksleri depola
        outlier_indices.extend(outlier_list_col)
    
    # esiz aykr deerleri bul
    outlier_indices = Counter(outlier_indices)
    # eer bir rnek (v) 1 farkl stun iin aykr deerse bunu aykr olarak kabul et (v>1)
    multiple_outliers = list(i for i, v in outlier_indices.items() if v > 1)
    
    return multiple_outliers
```

```python
veri_anomali = veri.loc[anomaliTespiti(veri,["yas","kilo","boy"])]
veri_anomali.spor.value_counts()
#buraya kadar ----
```
