# Veri Görselleştirme
* **Veri Görselleştirme Nedir?**
    * Bugün iş dünyasında veri analizi yoluyla toplanan bu kadar çok bilgi ile, bu verilerin bir resmini çizmenin bir yolunu bulmalıyız, böylece onları yorumlayabiliriz. 
    * Veri görselleştirme, haritalara veya grafiklere görsel bağlam sağlayarak bilginin ne anlama geldiğine dair bize net bir fikir verir. 
    * Bu, verileri insan zihninin anlaması için daha doğal hale getirir ve bu nedenle büyük veri kümelerindeki eğilimleri, modelleri ve aykırı değerleri tanımlamayı kolaylaştırır.
    * Kısaca, veri görselleştirme verilerin veya bilgilerin bir grafik, çizelge veya diğer görsel formatta temsilidir. 
* **Veri Görselleştirmenin Önemi**
    * Ana amaç, veri ve istatistikleri görselleştirmek, bilgi edinmek için ekranları yorumlamaktır. 
    * Veri görselleştirme, veri temizleme, veri yapısını keşfetme, aykırı değerleri ve olağan dışı grupları tespit etme, eğilimleri ve kümeleri belirleme, yerel kalıpları belirleme, modelleme çıktısını değerlendirme ve sonuçları sunma için kullanışlıdır.
    * Makine öğrenimi, tahmine dayalı analiz gibi analizler yapmayı kolaylaştırır ve bu analizler, daha sonra sunum için yararlı görselleştirmeler işlevi görebilir.
    * Finans, pazarlama, teknoloji, tasarım veya başka herhangi bir alanda çalışıyor olun, verileri görselleştirmeniz gerekir. Bu gerçek, veri görselleştirmenin önemini göstermektedir.
* **Veri Görselleştirmede En Çok Kullanılan Python Kütüphaneleri**
    1. **Matplotlib:** Python temellerini hatırlama bölümünde gördüğümüz temel görselleştirme kütüphanesidir.
    1. **Seaborn:** İstatistiksel bir Python veri görselleştirme kütüphanesidir. (Bu bölümde göreceğiz)
    1. **Plotly:** Etkileşimli bir Python veri görselleştirme kütüphanesidir. (Bu bölümde göreceğiz)



## Seaborn Kütüphanesi
* Seaborn, Matplotlib kütüphanesi tabanlı, istatistiksel bir Python veri görselleştirme kütüphanesidir. 
* Seaborn kullanıcılara istatistiksel görselleştirmeler yapmaları için yüksek seviyeli bir arayüz sunar.
* Seaborn kütüphanesi’nin işlevleri aşağıdaki gibidir.
    * Birden çok değişken arasındaki ilişkileri detaylı bir şekilde incelemek adına sunulan veri kümesi tabanlı bir uygulama programlama arayüzüdür.
    * Kategorik değişkenleri gözlemlemek ve üzerinde istatistiksel işlemler yapmak için destek sağlar.
    * Karmaşık görselleştirmeler oluşturmayı sağlayan çoklu grafikleri destekler.
    * Alt kümeler arasında karşılaştırma yapmak için tek değişkenli ve çift değişkenli görselleştirmeler sunar.
    * Görselleştirmeyi daha anlaşılır hale getiren geniş renk skalası mevcuttur.
* Seaborn ve Matplotlib kütüphanesi arasındaki farklar:
    * Matplotlib kişiselleştirilebilir; fakat plot'ları daha anlaşılır, göze hitap eden şekle getirmek bir hayli zordur.
    * Seaborn sahip olduğu renk paleti, arayüzler ve grafik çeşitleri ile kullanıcıya farklı çözümler sunar.
    * Pandas Kütüphanesi ile çalışırken Seaborn kolaylık sağlar.

```python
# kütüphaneleri import edelim
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# pythonda uyarıları kapatır
import warnings
warnings.filterwarnings("ignore")
```

```python
# veriyi içe aktar
veri = pd.read_csv("olimpiyatlar_temizlenmis.csv")
veri.head()
```

## Saçılım Grafiği
*  Saçılım grafiği, iki farklı değişkenin arasındaki ilişkiyi belirlemek için kullanılır.

```python
# Saçılım grafiği çizmek için bir plot yapalım
def sacilimGrafik():
    sns.scatterplot(x='boy', y='kilo', data=veri)
    plt.title('Boy ve Kilo Dağılımı - Beyaz Izgara Tema')
    plt.show()
sacilimGrafik()    
```

```python
# seaborn stilleri
sns.set_style("white")
sacilimGrafik()
```

```python
# seaborn stilleri
sns.set_style("darkgrid")
sacilimGrafik()
```

```python
# seaborn stilleri ve dağılım gradiği 
# son olarak beyaz ızgara teması ile devam edelim
sns.set_style("whitegrid")
sacilimGrafik()
```

* Bu grafiklerden boy ve kilo arasında pozitif ilişki olduğu anlamını çıkarabiliriz.

<a id='3'></a>
## Saçılım Grafiği - Kategorik

```python
# hue = sınıf ya da etiket olarak adlandırılabilir.
# scatter plot çizdirilirken hue'ye göre sınıflar farklı renklendirili
sns.scatterplot(x='boy', y='kilo', hue='madalya', data=veri)
plt.title('Madalyaya Göre Boy ve Kilo Dağılımı')
plt.show()
```

<a id='4'></a>
## Saçılım Grafiği - Doğrusal Regresyon

```python
# saçılım grafiği yapılır üzerine doğrusal regresyon yöntemi ile line fit işlemi gerçekleştirilir
# doğrusal regresyon makine öğrenmesinde anlatılacaktır.
sns.regplot(x='boy', y='kilo', data=veri, marker="+", scatter_kws={'alpha':0.2})
plt.title('Boy ve Kilo Dağılımı')
plt.show()
```

* Doğrusal regresyon sonucu oluşan, doğrunun eğiminin pozitif olmasından anlaşılacağı üzere boy ve kilo arasında pozitif ilişki vardır.

```python
# buraya kod yazacağız
# kütüphaneleri import edelim
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# pythonda uyarıları kapatır
import warnings
warnings.filterwarnings("ignore")

# plotly 
from plotly.offline import init_notebook_mode, iplot, plot
import plotly as py
init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.io as pio
import plotly.figure_factory as ff
```

```python
# buraya kod yazacağız
veri = pd.read_csv("universite_siralamasi.csv")
veri.head(3)
```

```python
# buraya kod yazacağız
veri.columns
```

```python
# buraya kod yazacağız
# sütun isimlerinin değiştirilmesi
veri.rename(columns={'world_rank'             : 'dunya_siralama', 
                     'university_name'        : 'uni_isim', 
                     'country'                : 'ulke', 
                     'teaching'               : 'ogretim', 
                     'international'          : 'uluslararası', 
                     'research'               : 'arastirma', 
                     'citations'              : 'alinti', 
                     'income'                 : 'gelir', 
                     'total_score'            : 'toplam_puan',
                     'num_students'           : 'ogrenci_sayisi', 
                     'student_staff_ratio'    : 'ogrenci_calisan_orani', 
                     'international_students' : 'uluslararası_ogrenci',
                     'female_male_ratio'      : 'kadin_erkek_orani',
                     'year'                   : 'yil'}, inplace=True) # inplace = True dediğimiz zaman ismi değiştirilen veri otomatik olarak veri variable'a kaydedilir
veri.head(2)
```

## Çizgi Grafiği
* Çizgi grafiği, birbirine bağlanan düz çizgi parçalarının veri noktaları serisinin bilgisini gösteren bir tür çizimdir. 
* Çizelge genellikle belirli bir zaman aralığındaki - zaman serisi - verilerin değerlerini görselleştirmek için kullanılır ve kronolojiksel olarak çizilir. 
* Ayrıca zamana bağlı ölçümlerin görselleştirilmesi için de kullanılır.

```python
# buraya kod yazacağız
# ilk 100 örneği al
df = veri.iloc[:100,:]
df.head(3)
```

```python
# buraya kod yazacağız
# plotly stilleri 
pio.templates.default = "simple_white" # grafik stilini beyaz yap
```

```python
# buraya kod yazacağız
# ilk grafik
cizgi1 = go.Scatter(x = df.dunya_siralama,  # x ekseninde dünya sıralaması olsun
                    y = df.alinti,          # y ekseninde alıntı olsun
                    mode = "lines+markers", # grafik çizgi ve saçılım olsun
                    name = "Alıntı",        # etiketi alıntı olsun
                    marker = dict(color = 'rgba(78, 78, 250, 0.85)'), # renk
                    text= df.uni_isim)      # grafik üzerinde imgeç ilgili nokyata gelince üniversite ismi yazsın
# ikinci grafik
cizgi2 = go.Scatter(x = df.dunya_siralama, # x ekseninde dünya sıralaması olsun
                    y = df.ogretim,        # y ekseninde öğretim olsun
                    mode = "lines+markers",# grafik çizgi ve saçılım olsun
                    name = "Öğretim",      # etiketi öğretim olsun
                    marker = dict(color = 'rgba(202, 43, 15, 0.85)'), # renk
                    text= df.uni_isim)     # grafik üzerinde imgeç ilgili nokyata gelince üniversite ismi yazsın
veri_ = [cizgi1, cizgi2] # 2 grafiği veri içerisine koy
# başlık ekle ve eksen ayarlarını yap
yerlesim = dict(title = 'Dünya Sıralamasında İlk 100 Üniversitenin Alıntı ve Öğretim Puanları',
              xaxis= dict(title= 'Dünya Sıralaması',ticklen= 5,zeroline= False))
fig = dict(data = veri_, layout = yerlesim) # figür oluştur 
plot(fig, filename = '1_çizgi-alıntı ve öğretim puanları.html') # .html olarak kaydet
```

```python
# buraya kod yazacağız
# 2014, 2015 ve 2016 yıllarını ayıralım ve ilk 100'er örneği alalım
veri2014 = veri[veri.yil == 2014].iloc[:100,:]
veri2015 = veri[veri.yil == 2015].iloc[:100,:]
veri2016 = veri[veri.yil == 2016].iloc[:100,:]

# ilk grafik
cizgi1 =go.Scatter( x = veri2014.dunya_siralama, # x ekseninde dünya sıralaması olsun
                    y = veri2014.alinti,         # y ekseninde alıntı olsun
                    mode = "markers",            # grafik saçılım olsun
                    name = "2014",               # etiketi 2014 olsun
                    marker = dict(color = 'rgba(2, 255, 15, 0.8)'), # renk
                    text= veri2014.uni_isim)    # grafik üzerinde imgeç ilgili nokyata gelince üniversite ismi yazsın
# ikinci grafik
cizgi2 =go.Scatter( x = veri2015.dunya_siralama,
                    y = veri2015.alinti,
                    mode = "markers",
                    name = "2015",
                    marker = dict(color = 'rgba(255, 0, 123, 0.8)'),
                    text= veri2015.uni_isim)
# üçüncü grafik
cizgi3 =go.Scatter( x = veri2016.dunya_siralama,
                    y = veri2016.alinti,
                    mode = "markers",
                    name = "2016",
                    marker = dict(color = 'rgba(0, 5, 200, 0.8)'),
                    text= veri2016.uni_isim)
veri_ = [cizgi1, cizgi2, cizgi3]
yerlesim = dict(title = '2014, 2015 ve 2016 yıllarında ilk 100 üniversinin alıntı ve dünya sıralaması',
              xaxis= dict(title= 'Dünya Sıralaması',ticklen= 5,zeroline= False),
              yaxis= dict(title= 'Alıntı',ticklen= 5,zeroline= False)
             )
fig = dict(data = veri_, layout = yerlesim)
plot(fig, filename = '2_saçılım-alıntı ve dünya sıralaması.html')
```

## Çubuk Grafiği
* Çubuk grafik veya çubuk grafik, kategorik verileri temsil ettikleri değerlerle orantılı yükseklik veya uzunluktaki dikdörtgen çubuklarla sunan bir grafiktir. 

```python
# buraya kod yazacağız
# 2014 yılının ilk 5 üniversitesini alalım
veri2014 = veri[veri.yil == 2014].iloc[:5,:]

# ilk grafik
cizgi1 = go.Bar(x = veri2014.uni_isim,
                y = veri2014.alinti,
                name = "Alıntı",
                marker = dict(color = 'rgba(255,127, 39, 0.5)', # bar rengi
                             line=dict(color='rgb(0,0,0)',width=1.5)), # bar'ın etrafındaki çizgilerin rengi ve kalınlığı
                text = veri2014.ulke)
# ikinci grafik
cizgi2 = go.Bar(x = veri2014.uni_isim,
                y = veri2014.ogretim,
                name = "Öğretim",
                marker = dict(color = 'rgba(64, 0, 128, 0.5)',
                              line=dict(color='rgb(0,0,0)',width=1.5)),
                text = veri2014.ulke)
veri_ = [cizgi1, cizgi2]
yerlesim = go.Layout(barmode = "group")
fig = go.Figure(data = veri_, layout = yerlesim)
plot(fig,filename = '3_çubuk-alıntı ve öğretim puanı.html')
```

## Dairesel Grafik
* Dairesel grafik, istatistik biliminde betimsel istatistik alanında kategorik verileri görsel bir şekilde betimleyip özetlemek için hazırlanan; içindeki kategori dilimlerini orantısal olarak gösteren bir daire şeklinde sunulan bir gösterim aracıdır.

```python
# buraya kod yazacağız
# 2016 yılının ilk 8 üniversitesini alalım
veri2016 = veri[veri.yil == 2016].iloc[:8,:]
dilim1 = veri2016.ogrenci_sayisi
dilim1_liste = [float(each.replace(',', '.')) for each in veri2016.ogrenci_sayisi]  # str(2,4) => str(2.4) = > float(2.4) = 2.4
etiketler = veri2016.uni_isim

# ilk grafik
cizgi = go.Pie(labels = etiketler, 
               values = dilim1_liste,
               hoverinfo = 'label+value+percent', 
               textinfo = 'value+percent', 
               textfont = dict(size=8),
               rotation = 180,
               hole = 0.3,
               marker = dict(line=dict(color='#000000', width=1)))

veri_ = [cizgi]
yerlesim = dict(title = "2016 yılında dünya sıralamasında ilk 8 de bulunan üniversitelerin öğrenci sayıları ve oranı",
                legend=dict(orientation="h"))

fig = dict(data=veri_,layout=yerlesim)
plot(fig, filename = '4_dairesel-öğrenci sayısı.html') # .html olarak kaydet
```

## Histogram
* Histogram, gruplandırılmış bir veri dağılımının sütun grafiğiyle gösterimidir. 
* Tekrarlı sayılardan oluşan verilerin, uygulanan işlemlerden sonra önce tabloya, tablodan yararlanarak grafiğe aktarılması, yani veri gruplarının grafiğinin dikdörtgen sütunlar halinde gösterilmesidir.

```python
# buraya kod yazacağız
# 2011 ve 2012 yıllarında öğrenci/çalışan oranını alalım
veri2011 = veri.ogrenci_calisan_orani[veri.yil == 2011]
veri2012 = veri.ogrenci_calisan_orani[veri.yil == 2012]

# ilk grafik
cizgi1 = go.Histogram(x = veri2011,
                      opacity = 0.75,
                      name = "2011",
                      marker = dict(color='rgba(171, 50, 144, 0.6)'))
# ikinci grafik
cizgi2 = go.Histogram(x = veri2012,
                      opacity = 0.75,
                      name = "2012",
                      marker = dict(color='rgba(23, 150, 196, 0.6)'))

veri_ = [cizgi1, cizgi2]

yerlesim = go.Layout(barmode = 'overlay',
                   title = '2011 ve 2012 yıllarında öğrenci/çalışan oranı',
                   xaxis = dict(title='Öğrenci Çalışan Oranı'),
                   yaxis = dict( title='Frekans'),
)
fig = go.Figure(data=veri_, layout=yerlesim)
plot(fig, filename = '6_histogram-öğrenci çalışan oranı.html') # .html olarak kaydet
```

## Dünya Haritası Grafiği
* Dünya haritası grafiği ile, veri içerisinde bulunan ülke-kıta gibi özelliklerin ve bu özelliklere karşılık gelen sayısal verinin görselleştirilmesi sağlanır.
* Dünya haritası üzerinde görselleştirme yapılırken, aşağıda bulunan paletlerden biri seçilebilir.


```python
# 2016 senesini verisini alalım
veri2016 = veri[veri.yil == 2016]
veri2016.ulke.value_counts() 
```

```python
# amerikarada 146 adet okul olduğu için toplamı 146'ya böldüm, görselleştirme daha güzel görünsün diye
ulkeye_gore_toplam_veriler = veri2016.groupby("ulke").sum()/146
ulkeye_gore_toplam_veriler
```

```python
# ilk grafik
cizgi1 = go.Choropleth( locations = ulkeye_gore_toplam_veriler.index,
                       locationmode='country names',
                       z = ulkeye_gore_toplam_veriler['arastirma'],
                       text = ulkeye_gore_toplam_veriler.index,
                       autocolorscale =False,
                       reversescale = True,
                       colorscale = "iceFire",
                       marker = dict(line = dict(color = 'rgb(255,255,255)',width = 1)),
                                     colorbar = dict( title = 'Araştırma Puanı',tickprefix = ''))
veri_ = [cizgi1]
yerlesim = go.Layout( title = "Ülkelerin Toplam Araştırma Puanları",
                      geo = dict(showframe = True,
                      showlakes = False,
                      showcoastlines = True,
                      projection = dict( type = 'natural earth')))

fig = dict( data=veri_, layout=yerlesim )
plot(fig, filename = '10_dünya haritası- araştırma puanı.html')
```
